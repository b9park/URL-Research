# Geomapping API Function

```python
def get_lat_lng(row):
    
    address = row['college']
    # Setting up parameters for API
    params = {'key': API_KEY, 'address': address}
    base_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
    
    # Setting up API request
    response = requests.get(base_url, params=params).json()
    if response['status'] == 'OK':
            geo = response['results'][0]['geometry']
            latitude = geo['location']['lat']
            longitude = geo['location']['lng']
            return latitude, longitude
        
# Applying function to dataframe
coordinates = df.apply(get_lat_lng, axis=1)
df['new_latitude'] = coordinates.apply(lambda x: x[0])
df['new_longitude'] = coordinates.apply(lambda x: x[1])
```
